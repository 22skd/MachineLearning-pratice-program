```python
from sklearn.datasets import load_iris
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier
```

```python
print(load_iris())
iris_data=load_iris()
X=iris_data.data[:,2:]
print(X)
y=iris_data.target
```

```python
tree_clf=DecisionTreeClassifier(criterion="entropy",max_depth=4,random_state=42)
clf=tree_clf.fit(X,y)

from sklearn import tree
tree.plot_tree(clf,feature_names=iris_data.feature_names[2:],class_names=iris_data.target_names)
```

```python
print(tree_clf.predict_proba([[5,1.5]]))
otp=tree_clf.predict([[5,1.5]])
print(iris_data.target_names[otp])
```
